```python
def read_images(image_path):
  read_image = cv2.imread(image_path)
  return read_image

def encode_faces(input_image):
  return face_recognition.face_encodings(input_image)[0]

def find_distance(input_image1,input_image2):
  img_enc1 = encode_faces(input_image1) # encoding first image
  img_enc2 = encode_faces(input_image2) # encoding second image
  distance = face_recognition.face_distance([img_enc1], img_enc2) # finding the distance
  return distance[0]

def detect_face(threshhold,distance_between_faces):
  if distance_between_faces > threshhold:
    return (False,"Faces are not matching, Matching score: ",((1 - distance_between_faces)*100),"%")
  else:
    return (True,"Faces matched, Matching score: ",((1 - distance_between_faces)*100),"%")

def comare_two_images(imagepath1,imagepath2):
  image_raw1 = read_images(imagepath1)
  image_raw2 = read_images(imagepath2)
  face_distance_cal = find_distance(image_raw1,image_raw2)
  print(detect_face(0.5,face_distance_cal))
```

```python
def get_detected_face(input_image):
  gray_image = cv2.cvtColor(input_image, cv2.COLOR_BGR2GRAY)
  face_classifier = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
  face = face_classifier.detectMultiScale(gray_image, scaleFactor=1.1, minNeighbors=5, minSize=(40, 40))
  img_cropped = input_image[face[0][1]:face[0][1]+face[0][3],face[0][0]:face[0][0]+face[0][2]]
  #img_cropped = cv2.cvtColor(img_cropped, cv2.COLOR_BGR2RGB)
  return img_cropped
```
